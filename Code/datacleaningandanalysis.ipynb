```python
def descriptive_analysis(df):
    average_rating = df["rating"].mean()
    rating_distribution = df["rating"].value_counts().sort_index()

    plt.figure(figsize=(8, 5))
    plt.hist(df["rating"], bins=10, edgecolor='black', alpha=0.7)
    plt.xlabel("Rating")
    plt.ylabel("Frequency")
    plt.title("Distribution of Ratings")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

    most_reviewed_products = df.nlargest(5, "reviews_count")[["title", "reviews_count"]]


    average_price = df["item_price"].mean()
    min_price = df["item_price"].min()
    max_price = df["item_price"].max()

    return {
        "average_rating": average_rating,
        "rating_distribution": rating_distribution,
        "most_reviewed_products": most_reviewed_products,
        "price_statistics": (average_price, min_price, max_price)
    }
```

```python
categories = {
    "iPhone": "iPhone",
    "MacBook": "MacBook",
    "iPad": "iPad",
    "Watch": "Watch",
    "AirPods": "AirPods"
}
```

```python
def assign_category(title):
    for category, keyword in categories.items():
        if keyword.lower() in title.lower():
            return category
    return "Other" 
```

```python
apple_df["product_category"] = apple_df["title"].apply(assign_category)
```

```python
apple_df.shape
```

```python
apple_df[["title", "product_category"]].head()
```

```python
category_summary = apple_df.groupby("product_category").agg(
    avg_price=("item_price", "mean"),
    avg_rating=("rating", "mean"),
    avg_reviews=("reviews_count", "mean")
).round(2)
print(category_summary)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
category_summary["avg_price"].plot(kind="bar", ax=ax[0], color="skyblue")
ax[0].set_title("Average Price by Product Category")
ax[0].set_ylabel("Price ($)")
category_summary["avg_rating"].plot(kind="bar", ax=ax[1], color="lightcoral")
ax[1].set_title("Average Rating by Product Category")
ax[1].set_ylabel("Rating (out of 5)")
category_summary["avg_reviews"].plot(kind="bar", ax=ax[2], color="lightgreen")
ax[2].set_title("Average Reviews Count by Product Category")
ax[2].set_ylabel("Number of Reviews")

for a in ax:
    a.set_xticklabels(a.get_xticklabels(), rotation=45)
plt.tight_layout()
plt.show()
```

Checking if Certain Models Receive More Negative Reviews


```python
import json
df_iphone["amount_of_stars"] = df_iphone["amount_of_stars"].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
```

```python
# Expand dictionary into new columns
df_stars = df_iphone["amount_of_stars"].apply(pd.Series)
df_stars.columns = ["1_star", "2_stars", "3_stars", "4_stars", "5_stars"]
df_iphone = pd.concat([df_iphone, df_stars], axis=1)
```

```python
df_iphone["negative_reviews"] = df_iphone["1_star"] + df_iphone["2_stars"]
df_iphone.groupby("title")["negative_reviews"].sum().sort_values(ascending=False).head(20)
```
