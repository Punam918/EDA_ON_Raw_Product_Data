```python
import pandas as pd
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'\s+', ' ', text)  # Normalize spaces
    text = re.sub(r'[^a-zA-Z0-9,.!$%&-]', ' ', text)  # Keep essential symbols
    text = text.strip()  # Remove leading/trailing spaces
    return text

def expand_abbreviations(text):
    ABBREVIATIONS = {
        "w/": "with",
        "qty": "quantity",
        "pkg": "package",
    }
    for abbr, full in ABBREVIATIONS.items():
        text = text.replace(abbr, full)
    return text

def remove_stopwords(text):
    return " ".join(word for word in text.split() if word not in stop_words)

def preprocess_text(text):
    text = clean_text(text)
    text = expand_abbreviations(text)
    text = remove_stopwords(text)  # Optional
    return text
```

```python
new_data['cleaned_product_description'] = new_data['product_description'].apply(preprocess_text)
new_data['cleaned_product_details'] = new_data['product_details'].apply(preprocess_text)

print(new_data[['cleaned_product_description', 'cleaned_product_details']])
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def get_sentiment(text):
    if not isinstance(text, str) or text.strip() == "":
        return 0  # Neutral for empty text
    scores = analyzer.polarity_scores(text)
    return scores['compound']  
```

```python
# Apply sentiment analysis
new_data['sentiment_product_description'] = new_data['cleaned_product_description'].apply(get_sentiment)
new_data['sentiment_product_details'] = new_data['cleaned_product_details'].apply(get_sentiment)
```
